###**import nedded library** 

```python
import cv2  
import matplotlib.pyplot as plt  # Is used to drow 2D plot
import easyocr  # Text Reader
from pylab import rcParams  # You can dynamically change the default rc (runtime configuration) 
from IPython.display import Image   # Used for image perprocessing
```

```python
rcParams['figure.figsize'] = 8 , 16
```

 ### **Import Image from file to read the text in image using easyocr**

```python
file_Name = '/content/download.png'   # Image Path
#Image(file_Name)
```

```python
img = cv2.imread(file_Name)   # Read the image Using imread()
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

###======================== **EasyOCR Model start** ================================

####***In easyocr the Reader() in used to read the text***

####**"en" means English Text & "hi" means Hindi Text** 

```python
reader = easyocr.Reader(["en","hi"])  # Create function for read the text ("reader")
```

```python
Output = reader.readtext(img) # Read Text From Image (extract text from a given image)
```

```python
Output #Display Output (Text)
```

###========================= **END** ==================================

##**Text dectection Using Index Value**

```python
cord = Output[2][0]   # Get index value
```

```python
x_min ,y_min = [int(min(idx)) for idx in zip(*cord)] # Get min idx value 
```

```python
x_max,y_max = [int(max(idx)) for idx in zip(*cord)]  # Get max inx value
```

```python
image = cv2.imread(file_Name) 
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(255,0,0),2)  # drow Rectangle in particular text using this function
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))  # Display 
```
